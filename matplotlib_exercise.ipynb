## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_hist_data = pd.read_csv("data/hist_data.csv")
plt.hist(df_hist_data["aデータ"], bins = 30, color = "yellow", alpha = 0.5, label = "aデータ")   #alphaは透明度をあわわす
plt.hist(df_hist_data["bデータ"], bins = 30, color = "cyan", alpha = 0.5, label = "bデータ") 
plt.xlabel("温度(℃）")
plt.ylabel("度数")
plt.legend(bbox_to_anchor = (1,1), loc = "upper right", borderaxespad = 1)
plt.title("aデータとbデータの温度に関する度数分布")
```

```python
temperature = pd.read_csv("data/hist_data.csv")


plt.title("aデータとbデータの温度に関する度数分布",  fontsize = 12)
plt.ylabel("度数", fontsize = 10)
plt.xlabel("温度(℃)", fontsize = 10)
plt.grid(False)
plt.hist(temperature["aデータ"], bins=30, alpha=0.6, label=["aデータ"], color="orange")
plt.hist(temperature["bデータ"], bins=30, alpha=0.6, label=["bデータ"], color="c")

plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_scatter = pd.read_csv("data/scatter_data.csv")
df_class0 = df_scatter[df_scatter["class"] == 0]
df_class1 = df_scatter[df_scatter["class"] == 1]
df_class2 = df_scatter[df_scatter["class"] == 2]

label = df_scatter.columns

plt.scatter(df_class0["length (cm)"], df_class0["width (cm)"], color = "red", label = "class0")
plt.scatter(df_class1["length (cm)"], df_class1["width (cm)"], color = "blue", label = "class1")
plt.scatter(df_class2["length (cm)"], df_class2["width (cm)"], color = "green", label = "class2")


plt.title("クラス0〜2のlengthとwidthに関する散布図")

plt.xlabel(label[0])
plt.ylabel(label[1])
plt.grid()
plt.legend()

plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

```python
df_line_data = pd.read_csv("data/line_data.csv")
plt.plot(df_line_data["期間"], df_line_data["件数"], marker = "o")  #maekerは線と線をつなぐ点を指定できる

plt.xlabel("期間")
plt.ylabel("件数")

plt.grid(True)
plt.xticks(rotation = 30)

plt.title("期間ごとの件数の推移")
```

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
plt.bar(df_line_data["期間"], df_line_data["費用"])

plt.xlabel("期間")
plt.ylabel("費用（円）")

plt.grid(True)
plt.xticks(rotation = 30)

plt.title("期間ごとの費用の推移")

```

```python
df_line = pd.read_csv("data/line_data.csv")
df_line ["期間"] = ["2016下半期", "2017上半期", "2017下半期", "2018上半期", "2018下半期", "2019上半期", "2019下半期", "2010上半期"]


plt.plot(df_line["期間"], df_line["件数"], market = ".")
plt.xticks(df_line["期間"], rotation = 30)

plt.title("期間ごとの費用の推移")
plt.xlabel("期間")
plt.ylabel("費用（円）")

plt.grid()
plt.show()
```
